```python
import cv2
import numpy as np
import math
import os

# Create our body classifier
body_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')


# Initiate video capture for video file, here we are using the video file in which pedestrians would be detected
cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_SIMPLEX
i = 65

# Loop once video is successfully loaded
while cap.isOpened():

    # Reading the each frame of the video 
    ret, frame = cap.read()

    if ret:
        centroid_x = []
        centroid_y = []
        # Pass frame to our body classifier as gray for better results
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        bodies = body_classifier.detectMultiScale(gray)
        if (len(bodies) != 0):
            num_of_bodies = len(bodies)
            
            # Extract bounding boxes for any bodies identified
            for (x,y,w,h) in bodies:
                cv2.putText(frame, str(num_of_bodies), (10,20),font, 0.5, (0,255,255), 2)#print number of bodies detected at corner
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 1)#draw rectangles
                crop_img = frame[ y:y+h,x:x+w]
                centroid_x.append((int((x + (x+w))/2)))#store all x coordinates of centroids in array
                centroid_y.append((int((y + (y+h))/2)))#store all y coordinates of centroids in array
            
            
            #print processed frame
            cv2.imshow('Pedestrians', crop_img)
            i = i+1
            cv2.imwrite(os.path.join('D:/jupyter/keras/Adv_CNN/faces/train/n02' ,(str(i)+'.jpg')), crop_img)
            
        else:
            #print non-processed frame
            cv2.imshow('Pedestrians', frame)
    else:  
       print('no video')
       cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
    
    if cv2.waitKey(10) == 13: #13 is the Enter Key
        break

cap.release()
cv2.destroyAllWindows()
```
